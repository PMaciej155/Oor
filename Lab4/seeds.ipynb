```python
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import numpy as np 
import pandas as pd 
```

```python
data = pd.read_csv("new_seeds.txt", delimiter=",", header=None, names=["area","perimeter", "compactness","length","width", "asymconf", "lengrov", "type"])
data
```

```python
plt.scatter(data.iloc[:, 5], 
            data.iloc[:, 6], 
            c='blue',
            marker='o', 
            s=50)
plt.grid()
plt.show()
```

```python
km = KMeans(n_clusters=3, 
            init='random', 
            n_init=10, 
            max_iter=2, 
            tol=1e-04,
            random_state=0)
y_km = km.fit_predict(data.iloc[:, [4, 6]])
```

```python

plt.scatter(data.iloc[y_km==0,5], 
            data.iloc[y_km==0,6], 
            s=50, 
            c='lightgreen', 
            marker='s', 
            label='cluster 1')

plt.scatter(data.iloc[y_km==1,5], 
            data.iloc[y_km==1,6], 
            s=50, 
            c='orange', 
            marker='o', 
            label='cluster 2')
plt.scatter(data.iloc[y_km==2,5], 
            data.iloc[y_km==2,6], 
            s=50, 
            c='lightblue', 
            marker='v', 
            label='cluster 3')
"""      
plt.scatter(data.iloc[:, 5], 
            data.iloc[:, 6], 
            c='blue',
            marker='o', 
            s=50)
"""     
plt.scatter(km.cluster_centers_[:,0],
            km.cluster_centers_[:,1], 
            s=250, marker='*', 
            c='red', 
            label='centroids')
plt.legend()
plt.grid()
plt.show()
```
